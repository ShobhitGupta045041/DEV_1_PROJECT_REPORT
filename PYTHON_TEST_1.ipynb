```python
#Task 1
word = "Anaconda"
word = word.lower()  # Convert the word to lowercase to remove case sensitivity

letter_count = {}  # Initialize an empty dictionary to store letter frequencies

for letter in word:
    if letter.isalpha():  # Check if the character is a letter
        if letter in letter_count:
            letter_count[letter] += 1
        else:
            letter_count[letter] = 1

# Calculate the number of unique letters
unique_letters_count = len(letter_count)

# Print the results
print(f"Number of Unique Letters = {unique_letters_count}")
print(letter_count)
```

```python
#Task 2
import random

# Generate random scores for each component
class_test_score = random.randint(10, 90)
mid_term_score = random.randint(10, 90)
project_score = random.randint(10, 90)
end_term_score = random.randint(10, 90)

# Calculate the weighted evaluation score
evaluation_score = (
    0.10 * class_test_score
    + 0.20 * mid_term_score
    + 0.30 * project_score
    + 0.40 * end_term_score
)

# Round off the evaluation score to the nearest integer
evaluation_score = round(evaluation_score)

# Print the random scores and the evaluation score
print(f"Class Test Score: {class_test_score}")
print(f"Mid Term Score: {mid_term_score}")
print(f"Project Score: {project_score}")
print(f"End Term Score: {end_term_score}")
print(f"Evaluation Score (Rounded): {evaluation_score}")
```

```python
#Task 5
import random

# Generate a random number between 1 and 99
random_number = random.randint(1, 99)

# Initialize lists for even, odd, and prime numbers
even_numbers = []
odd_numbers = []
prime_numbers = []

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Generate lists of even, odd, and prime numbers up to the random number
for num in range(1, random_number + 1):
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)
    if is_prime(num):
        prime_numbers.append(num)

# Print the random number and the lists
print(f"The random number: {random_number}")
print(f"List of even numbers up to the random number: {even_numbers}")
print(f"List of odd numbers up to the random number: {odd_numbers}")
print(f"List of prime numbers up to the random number: {prime_numbers}")
```

```python
#Task 8
from datetime import datetime

# Input date in YYYY-MM-DD format (e.g., "2023-08-20")
input_date_str = input("Enter a date (YYYY-MM-DD): ")

# Convert the input string to a datetime object
input_date = datetime.strptime(input_date_str, "%Y-%m-%d")

# Get today's date
today = datetime.today()

# Calculate the difference between the two dates
date_difference = today - input_date

# Extract years, months, and days from the difference
years = date_difference.days // 365
months = (date_difference.days % 365) // 30
days = (date_difference.days % 365) % 30

# Print the result
print(f"Years: {years}")
print(f"Months: {months}")
print(f"Days: {days}")
```

```python
#Task 9
def print_diamond_pattern(n):
    for i in range(n):
        print(" " * (n - i - 1) + "* " * (i + 1))
    for i in range(n - 2, -1, -1):
        print(" " * (n - i - 1) + "* " * (i + 1))

size = 5  # Adjust the size as needed
print_diamond_pattern(size)
```

```python
import random

# Create List A with 10 random letters
list_a = [chr(random.randint(65, 90)) for _ in range(10)]  # Generates uppercase letters (A-Z)

# Create a copy of List A as List B
list_b = list(list_a)

# Shuffle both List A and List B
random.shuffle(list_a)
random.shuffle(list_b)

# Create dictionaries to map connections from A to B and vice versa
connections_a_to_b = dict(zip(list_a, list_b))
connections_b_to_a = dict(zip(list_b, list_a))

# Function to find connected paths
def find_connected_paths(element):
    connected_paths = []
    current_element = element
    while current_element in connections_a_to_b:
        connected_paths.append(current_element)
        current_element = connections_a_to_b[current_element]
    return connected_paths

# Print the mapping and connected paths
for element in list_a:
    connected_paths_a_to_b = find_connected_paths(element)
    connected_paths_b_to_a = []
    if element in connections_b_to_a:
        connected_paths_b_to_a = find_connected_paths(connections_b_to_a[element])
    print(f"{element} is connected to: {' -> '.join(connected_paths_a_to_b)}")
    print(f"   and from: {' -> '.join(connected_paths_b_to_a)}")
```
